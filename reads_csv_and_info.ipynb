```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import shutil

# Base folder
DATA_DIR = Path(r"C:\static_inference\data")
timestamp = datetime.utcnow().strftime("%Y%m%dT%H%M%SZ")

# Helper to safely write CSV with backup
def safe_overwrite(df: pd.DataFrame, file_path: Path):
    backup_path = file_path.with_suffix(file_path.suffix + f".bak.{timestamp}")
    shutil.copy2(file_path, backup_path)
    print(f"  🔐 Backup created: {backup_path.name}")

    tmp_path = file_path.with_suffix(".tmp.csv")
    df.to_csv(tmp_path, index=False)
    tmp_path.replace(file_path)
    print(f"  ✅ Updated file written: {file_path.name}")

# Loop through all CSVs
for csv_file in sorted(DATA_DIR.glob("*.csv")):
    print("=" * 100)
    print(f"📄 Processing: {csv_file.name}")

    try:
        df = pd.read_csv(csv_file, low_memory=False)
    except Exception as e:
        print(f"❌ Failed to read {csv_file.name}: {e}")
        continue

    # Detect int64 cols (excluding timestamp)
    int_cols = [c for c in df.columns if df[c].dtype == 'int64' and c.lower() != "timestamp"]

    if not int_cols:
        print("  ℹ️ No integer columns to convert.")
        continue

    # Convert int64 -> float64
    for col in int_cols:
        df[col] = df[col].astype(float)

    # Show summary of changes
    print(f"  🔄 Converted to float64: {int_cols}")

    # Overwrite safely
    safe_overwrite(df, csv_file)

print("\n✅ All applicable integer columns converted to float64 (except timestamp).")
```

```python
# Paste this single cell into your Jupyter notebook and run it on your machine.
import pandas as pd
from pathlib import Path
import json
import re
from datetime import datetime

# ---- CONFIG ----
DATA_DIR = Path(r"C:\static_inference\data")
OUT_JSON = DATA_DIR / "features.json"
OUT_CSV = DATA_DIR / "features.csv"
# ----------------

# Ensure data directory exists
DATA_DIR.mkdir(parents=True, exist_ok=True)

# Heuristics & helpers
def is_timestamp_series(s: pd.Series) -> bool:
    name = s.name.lower()
    if name == "timestamp" or "time" in name:
        non_null = s.dropna().astype(str)
        if non_null.empty:
            return False
        sample = non_null.head(200)
        parsed = 0
        for v in sample:
            try:
                pd.to_datetime(v)
                parsed += 1
            except Exception:
                pass
        return parsed / len(sample) > 0.7
    return False

def detect_boolean(s: pd.Series) -> bool:
    non_null = s.dropna().astype(str).str.strip().str.lower()
    if non_null.empty:
        return False
    unique_vals = set(non_null.unique())
    bool_like = {'true','false','0','1','yes','no','y','n'}
    if unique_vals.issubset(bool_like):
        return True
    if len(unique_vals) == 2 and all(len(x) <= 3 for x in unique_vals):
        return True
    return False

def detect_integer(s: pd.Series) -> bool:
    non_null = s.dropna().astype(str)
    if non_null.empty:
        return False
    coerced = pd.to_numeric(non_null, errors='coerce')
    if coerced.isna().sum() / len(coerced) > 0.2:
        return False
    return coerced.dropna().apply(float.is_integer).all()

def detect_float(s: pd.Series) -> bool:
    non_null = s.dropna().astype(str)
    if non_null.empty:
        return False
    coerced = pd.to_numeric(non_null, errors='coerce')
    if coerced.isna().sum() / len(coerced) > 0.2:
        return False
    return not coerced.dropna().apply(float.is_integer).all()

def suggest_unit_and_role(col_name: str):
    cn = col_name.lower()
    unit = None
    role = "feature"
    if any(k in cn for k in ["_pct", "percent", "percentage"]):
        unit = "percent"
    if any(k in cn for k in ["_kw", "kwh", "power", "power_kw"]):
        unit = "kW"
    if any(k in cn for k in ["_v", "voltage", "volt"]) and "voltage" in cn or cn.endswith("_v"):
        unit = "V"
    if any(k in cn for k in ["_ma", "_a", "current", "amp"]):
        unit = "A"
    if any(k in cn for k in ["_c", "temp", "temperature", "cel", "celsius"]):
        unit = "C"
    if any(k in cn for k in ["_km", "km_h", "kmh", "kph", "speed"]):
        unit = "km/h"
    if any(k in cn for k in ["_mph"]):
        unit = "mph"
    if any(k in cn for k in ["_psi"]):
        unit = "psi"
    if any(k in cn for k in ["_hz", "frequency", "vibration"]):
        unit = "Hz"
    if any(k in cn for k in ["distance", "odometer", "trip", "km"]):
        role = "distance_or_trip"
    if cn.startswith("is_") or cn.startswith("has_") or cn.startswith("flag_"):
        role = "boolean_flag"
    if cn in ("gps_lat", "gps_long", "latitude", "longitude") or "gps" in cn:
        unit = "degrees"
        role = "geolocation"
    if cn == "timestamp" or "time" in cn:
        role = "timestamp"
    return unit, role

# Find CSV files
csv_files = sorted(DATA_DIR.glob("*.csv"))

if not csv_files:
    print(f"No CSV files found in {DATA_DIR}. Drop your CSVs there and re-run this cell.")
else:
    features_dict = {
        "generated_at": datetime.utcnow().isoformat() + "Z",
        "source_folder": str(DATA_DIR),
        "files": {}
    }

    rows_for_csv = []

    for csv_file in csv_files:
        # Read as strings first (robust detection)
        try:
            df = pd.read_csv(csv_file, dtype=str, low_memory=False)
        except Exception as e:
            print(f"⚠️  Skipping {csv_file.name} — read error: {e}")
            continue

        file_meta = {
            "file_name": csv_file.name,
            "total_rows": int(len(df)),
            "columns": []
        }

        for col in df.columns:
            s = df[col]
            col_name = col  # already cleaned in your previous step
            non_null_count = int(s.dropna().shape[0])
            unique_count = int(s.nunique(dropna=True))
            sample_vals = s.dropna().head(5).tolist()

            # Type detection
            if is_timestamp_series(s) or col_name.lower() == "timestamp":
                inferred_type = "timestamp"
            elif detect_boolean(s):
                inferred_type = "boolean"
            elif detect_integer(s):
                inferred_type = "integer"
            elif detect_float(s):
                inferred_type = "float"
            else:
                inferred_type = "string"

            unit, role = suggest_unit_and_role(col_name)

            col_meta = {
                "name": col_name,
                "inferred_type": inferred_type,
                "nullable": bool(s.isna().any() or (s.astype(str).str.strip() == "").any()),
                "unique_count": unique_count,
                "sample_values": sample_vals,
                "unit": unit,
                "role": role
            }

            file_meta["columns"].append(col_meta)

            rows_for_csv.append({
                "file": csv_file.name,
                "column": col_name,
                "type": inferred_type,
                "nullable": col_meta["nullable"],
                "unique_count": unique_count,
                "sample_values": ";".join(map(str, sample_vals)),
                "unit": unit or "",
                "role": role
            })

        features_dict["files"][csv_file.name] = file_meta

    # Safe write (with backup)
    def backup_and_write(path: Path, bytes_content: bytes):
        if path.exists():
            bak = path.with_suffix(path.suffix + ".bak." + datetime.utcnow().strftime("%Y%m%dT%H%M%SZ"))
            path.rename(bak)
            print(f"Backup created: {bak.name}")
        path.write_bytes(bytes_content)

    json_bytes = json.dumps(features_dict, indent=2).encode("utf-8")
    backup_and_write(OUT_JSON, json_bytes)
    print(f"Wrote JSON to: {OUT_JSON}")

    df_rows = pd.DataFrame(rows_for_csv)
    if OUT_CSV.exists():
        bakcsv = OUT_CSV.with_suffix(OUT_CSV.suffix + ".bak." + datetime.utcnow().strftime("%Y%m%dT%H%M%SZ"))
        OUT_CSV.rename(bakcsv)
        print(f"Backup created for CSV: {bakcsv.name}")
    df_rows.to_csv(OUT_CSV, index=False)
    print(f"Wrote CSV summary to: {OUT_CSV}")

    # Also print JSON to output for immediate copy/paste
    print("\n\n===== JSON OUTPUT (copy this into your features file if you want) =====\n")
    print(json.dumps(features_dict, indent=2))
```

```python
import pandas as pd
from pathlib import Path
import numpy as np
from itertools import combinations

DATA_DIR = Path(r"C:\static_inference\data")

# Read all CSV files into memory as strings (robust) - if very large, switch to chunked approach
csv_files = sorted(DATA_DIR.glob("*.csv"))
if not csv_files:
    raise SystemExit(f"No CSV files found in {DATA_DIR}")

# Load dataframes (as strings to preserve everything), keep index order
dfs = {}
for p in csv_files:
    try:
        dfs[p.name] = pd.read_csv(p, dtype=str, low_memory=False)
    except Exception as e:
        print(f"Failed to read {p.name}: {e}")

def normalize_series_for_compare(s: pd.Series) -> pd.Series:
    """Normalize a series: strip whitespace, convert empty strings to NaN, keep as str."""
    s_str = s.astype(str).str.strip()
    # pandas will produce "nan" string for NaNs after astype(str) — treat those as missing
    s_str = s_str.replace(r'^(nan|none|NaN|None|\s*)$', np.nan, regex=True)
    return s_str

def compare_two_series(s1: pd.Series, s2: pd.Series):
    """
    Compare two series (aligned by position) and return:
      total, equal_count, percent_equal, all_equal, diffs_sample (list of tuples (idx,v1,v2))
    Comparison strategy:
      - Try numeric comparison where both sides are numeric-coercible
      - Else fallback to string comparison
      - Treat both NaN at same pos as equal
    """
    # Align lengths: compare up to min length
    n = min(len(s1), len(s2))
    s1 = s1.iloc[:n].reset_index(drop=True)
    s2 = s2.iloc[:n].reset_index(drop=True)

    # Normalize whitespace/empty -> NaN
    s1n = normalize_series_for_compare(s1)
    s2n = normalize_series_for_compare(s2)

    # Attempt numeric coercion
    s1_num = pd.to_numeric(s1n, errors='coerce')
    s2_num = pd.to_numeric(s2n, errors='coerce')

    # Positions where both are numeric (not NaN)
    both_numeric_mask = (~s1_num.isna()) & (~s2_num.isna())

    # Initialize equality boolean array
    eq = pd.Series(False, index=range(n), dtype=bool)

    # Numeric-equal positions: direct numeric equality (exact)
    if both_numeric_mask.any():
        eq[both_numeric_mask] = s1_num[both_numeric_mask].values == s2_num[both_numeric_mask].values

    # For positions not numeric-equal candidates, fallback to string compare (including NaN handling)
    fallback_mask = ~both_numeric_mask
    if fallback_mask.any():
        # For string compare, treat NaN as NaN: equal only if both NaN
        s1_f = s1n.fillna("<MISSING>")  # temporary sentinel
        s2_f = s2n.fillna("<MISSING>")
        eq[fallback_mask] = s1_f[fallback_mask].values == s2_f[fallback_mask].values

    # Now treat sentinel "<MISSING>" equality as true only when both were NaN (they are equal above)
    # Count equal
    equal_count = int(eq.sum())
    total = n
    percent_equal = 100.0 * equal_count / total if total > 0 else 0.0
    all_equal = (equal_count == total)

    # Collect small sample of differences for quick inspection (first up to 5)
    diffs_idx = np.where(~eq.values)[0]
    diffs_sample = []
    for idx in diffs_idx[:5]:
        v1 = s1n.iloc[idx]
        v2 = s2n.iloc[idx]
        diffs_sample.append((int(idx), v1 if pd.notna(v1) else None, v2 if pd.notna(v2) else None))

    return {
        "total_compared": total,
        "equal_count": equal_count,
        "percent_equal": percent_equal,
        "all_equal": all_equal,
        "diffs_sample": diffs_sample,
        "len_a": len(s1),
        "len_b": len(s2)
    }

# Now compare every pair of files
report_lines = []
for a_name, b_name in combinations(sorted(dfs.keys()), 2):
    df_a = dfs[a_name]
    df_b = dfs[b_name]
    common_cols = set(df_a.columns).intersection(df_b.columns)
    # Exclude any timestamp-like columns
    common_cols = [c for c in sorted(common_cols) if "timestamp" not in c.lower()]
    if not common_cols:
        print(f"\nNo common non-timestamp columns between {a_name} and {b_name}.")
        continue

    print("\n" + "="*100)
    print(f"Comparing files: {a_name}  ⟷  {b_name}")
    print(f"Rows: {a_name}={len(df_a)}, {b_name}={len(df_b)}")
    print(f"Common columns (excluding timestamp): {len(common_cols)} -> {common_cols}\n")

    for col in common_cols:
        s_a = df_a[col] if col in df_a.columns else pd.Series(dtype=str)
        s_b = df_b[col] if col in df_b.columns else pd.Series(dtype=str)
        res = compare_two_series(s_a, s_b)

        # Pretty print
        print(f" - Column: {col}")
        if res["len_a"] != res["len_b"]:
            print(f"    ⚠️  Length mismatch — compared up to min length {res['total_compared']} (A:{res['len_a']}, B:{res['len_b']})")
        print(f"    ✅  All equal: {res['all_equal']}  |  Equal: {res['equal_count']}/{res['total_compared']} ({res['percent_equal']:.2f}%)")

        if not res["all_equal"]:
            print(f"    🔍  Sample diffs (index, val_in_{a_name}, val_in_{b_name}):")
            for idx, va, vb in res["diffs_sample"]:
                print(f"         - [{idx}]  {va!r}  <>  {vb!r}")
        # small spacer
        print()
    print("="*100)

print("\nComparison complete.")
```

```python
# Rename conflicting columns in-place by appending _<module> to the column name in the second CSV of each conflicting pair.
import pandas as pd
from pathlib import Path
from itertools import combinations
from datetime import datetime
import shutil
import numpy as np

DATA_DIR = Path(r"C:\static_inference\data")
ts = datetime.utcnow().strftime("%Y%m%dT%H%M%SZ")

# Read CSVs
csv_paths = sorted(DATA_DIR.glob("*.csv"))
if not csv_paths:
    raise SystemExit(f"No CSV files found in {DATA_DIR}")

# load dataframes as strings (to preserve everything)
dfs = {}
for p in csv_paths:
    try:
        dfs[p.name] = pd.read_csv(p, dtype=str, low_memory=False)
    except Exception as e:
        print(f"Failed to read {p.name}: {e}")

def normalize_series_for_compare(s: pd.Series) -> pd.Series:
    s_str = s.astype(str).str.strip()
    s_str = s_str.replace(r'^(nan|none|NaN|None|\s*)$', np.nan, regex=True)
    return s_str

def series_all_equal(s1: pd.Series, s2: pd.Series) -> bool:
    """Return True if values (positional up to min length) are all equal under numeric-or-string compare."""
    n = min(len(s1), len(s2))
    if n == 0:
        return True
    a = normalize_series_for_compare(s1.iloc[:n].reset_index(drop=True))
    b = normalize_series_for_compare(s2.iloc[:n].reset_index(drop=True))
    # numeric compare where both numeric
    a_num = pd.to_numeric(a, errors='coerce')
    b_num = pd.to_numeric(b, errors='coerce')
    both_numeric = (~a_num.isna()) & (~b_num.isna())
    eq = pd.Series(False, index=range(n), dtype=bool)
    if both_numeric.any():
        eq[both_numeric] = a_num[both_numeric].values == b_num[both_numeric].values
    fallback = ~both_numeric
    if fallback.any():
        a_f = a.fillna("<M>")
        b_f = b.fillna("<M>")
        eq[fallback] = a_f[fallback].values == b_f[fallback].values
    return bool(eq.all())

# Track renames to apply per file (collect then apply to avoid multiple reads/writes mid-loop)
renames_by_file = {p.name: {} for p in csv_paths}  # mapping old_col -> new_col

# Compare every ordered pair (A,B) and plan renames in B
for a_name, b_name in combinations(sorted(dfs.keys()), 2):
    df_a = dfs[a_name]
    df_b = dfs[b_name]
    common = set(df_a.columns).intersection(df_b.columns)
    # exclude timestamp-like columns
    common = [c for c in sorted(common) if "timestamp" not in c.lower()]
    if not common:
        continue

    module_b = Path(b_name).stem  # e.g., 'tyre', 'transmission'
    for col in common:
        # If column already planned to be renamed in B earlier, refer to original column name for comparison
        b_col_current = col
        # Compare original A[col] with current B[col] (use original df from dfs - renames not yet applied)
        s_a = df_a[col]
        s_b = df_b[col]
        equal = series_all_equal(s_a, s_b)
        if not equal:
            # prepare new column name for B
            suffix = f"_{module_b}"
            if b_col_current.endswith(suffix):
                # already suffixed, skip
                continue
            candidate = b_col_current + suffix
            # Avoid collisions: if candidate exists or planned, add numeric suffix
            i = 1
            while (candidate in df_b.columns) or (candidate in renames_by_file[b_name].values()) or (candidate in renames_by_file[b_name].keys()):
                candidate = f"{b_col_current}{suffix}_{i}"
                i += 1
            renames_by_file[b_name][b_col_current] = candidate

# Apply renames in-place for each file that needs changes
summary = []
for fname, renames in renames_by_file.items():
    if not renames:
        continue
    path = DATA_DIR / fname
    print("="*80)
    print(f"Applying renames to {fname} (backup created)...")
    try:
        df = pd.read_csv(path, dtype=str, low_memory=False)
    except Exception as e:
        print(f"  ❌ Failed to read {fname} for writing: {e}")
        continue

    # Show planned renames
    for old, new in renames.items():
        if old not in df.columns:
            print(f"  ⚠️  Column {old} not present in current {fname}, skipping.")
            continue
        # perform rename
        df = df.rename(columns={old: new})
        print(f"  🔁 {old} -> {new}")
        summary.append((fname, old, new))

    # Backup original
    backup = path.with_suffix(path.suffix + f".bak.{ts}")
    shutil.copy2(path, backup)
    # Write temp then replace
    tmp = path.with_suffix(".tmp.csv")
    df.to_csv(tmp, index=False)
    tmp.replace(path)
    print(f"  ✅ Written updated file and backup saved as {backup.name}")

if not summary:
    print("No conflicting columns required renaming.")
else:
    print("\nSummary of renames performed:")
    for fname, old, new in summary:
        print(f" - {fname}: {old}  →  {new}")

print("\nDone. Sanity-check suggestion: run a columns-listing cell to verify final column names.")
```
